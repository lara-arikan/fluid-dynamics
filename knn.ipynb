```python
N = bo_df.shape[0] #number of rows
K = 417 #trial and error
```

```python
def euclidean_distance(point1, point2): 
    sum_squared_distance = 0
    for i in range(len(point1)):
        sum_squared_distance += math.pow(point1[i] - point2[i], 2)
    return math.sqrt(sum_squared_distance)

def mode(labels):
    return Counter(labels).most_common(1)[0][0]
```

```python
def df_to_tuples(dataframe):
    tuples_list = []
    lists = dataframe.values.tolist()
    for sublist in lists:
        tuples_list.append(tuple(sublist))
    return tuples_list

def knn(data, query, k): # data as list of tuples, query as tuple
    distances_and_indices = []
    
    for index, point in enumerate(data):
        assert len(point) == len(query), "Length mismatch in data points"
        distance = euclidean_distance(point, query)
        distances_and_indices.append((distance, index))

    k_nearest = sorted(distances_and_indices)[:k] #ascending order
    #print("Nearest", k_nearest)
    k_nearest_labels = [bo_df.iloc[i, 29] for distance, i in k_nearest]
    #print("Nearest labels:", k_nearest_labels)
    return mode(k_nearest_labels)
```

```python
queries = [tuple(bo_df.iloc[i, :9].values.tolist()) for i in range(100)]
```

```python
data = df_to_tuples(bo_df.iloc[101:, :9])
```

```python
accurate = 0
for i, query in enumerate(queries):
    if knn(data, query, K) == bo_df.iloc[i, 29]:
        accurate += 1
        
accuracy = accurate / len(queries)
accuracy
```
